1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

````python
from sklearn.model_selection import train_test_split

# Create an 80/20 split using train_test_split
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Report the number of observations in each set
print(f"Training set size: {ab_reduced_noNaN_train.shape[0]} observations")
print(f"Testing set size: {ab_reduced_noNaN_test.shape[0]} observations")
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_text
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']
X = ab_reduced_noNaN[['List Price']]

# Prepare the target variable (y) and the feature variable (X) for training
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Target: whether the book is hardcover
X_train = ab_reduced_noNaN_train[['List Price']]  # Feature: List Price

# Initialize the DecisionTreeClassifier with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Fit the classifier to the training data
clf.fit(X_train, y_train)

# Output the number of samples used to train the model
print(f"Model trained on {X_train.shape[0]} samples.")

# Optionally, print the decision tree rules (model structure)
tree_rules = export_text(clf, feature_names=["List Price"])
print(tree_rules)

# Now, let's evaluate the model's performance on the test set
# Prepare the target variable (y) and the feature variable (X) for testing
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # Target: whether the book is hardcover
X_test = ab_reduced_noNaN_test[['List Price']]  # Feature: List Price

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy on the test set: {accuracy * 100:.2f}%")

# Display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
cm_display.plot(cmap='Blues')

# Calculate Sensitivity (Recall)
sensitivity = recall_score(y_test, y_pred)
print(f"Sensitivity (Recall): {sensitivity:.4f}")

# Calculate Specificity
# Specificity = True Negative / (True Negative + False Positive)
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp)
print(f"Specificity: {specificity:.4f}")
````

````python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
plt.figure(figsize=(10, 8))
plot_tree(clf, feature_names=["List Price"], class_names=["Paper", "Hard"], filled=True, max_depth=2)
plt.title("Decision Tree Visualization")
plt.show()
````

The model classifies entries as Hard or Paper based solely on List Price. First, the model splits the books into List Price greater than 17.97 and less than/equal to 17.97. The left branch, representing "List Price less than/equal to 17.97" is classified as Paper. The right branch is split further, into less than/equal to 27.275 (Hard) and greater than 27.275 (Paper). The left branch is technically also split further, but the split is of no consequence to the final classification. 

Essentially, books with List Price greater than 17.97 and less than or equal to 27.275 are classified as Hard, otherwise they are classified as Paper.

````python
# Optionally, print the decision tree rules (model structure)
plt.figure(figsize=(10, 8))
plot_tree(clf2, feature_names=['List Price', 'NumPages', 'Thick'], class_names=["Paper", "Hard"], filled=True, max_depth=4)
plt.title("Decision Tree Visualization")
plt.show()
````

In this model, data is first split based on List Price; <=17.97 goes to the left branch and >17.97 to the right branch. The left branch is then split into NumPages; <= 82 to the left and > 82 to the right. This happens at most nodes (some end early - classification happens after level 3 and not level 4). Data are split based on their thickness, number of pages, and cost, in order to predict whether they are a hardcover or paperback.

Predictions are made with categorical splits, starting at the root node (first split). At each level (n = depth level), the data is categorized into 1 of 2 categories, depending on a true/false statement (e.g. List Price <= 19.79 or List Price > 19.79). Depending on the classification, the data then follows the left branch (True) or the right branch (False). The data is further categorized in a similar fashion at each consecutive level, with different conditions. Eventually, the data is classified based on the desired category (e.g. Hard/Paper).

The clf2 model splitting is optimized based on impurity metrics (e.g. Gini impurity, entropy), which basically measures how well a specific split will classify data. Higher impurity means that the split doesn't do much to categorize data (so for variables that don't really affect the category in question). Low impurity means that the split is effective at separating data.

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).

````python
# Extract feature importances
feature_importances = clf2.feature_importances_

# Pair the importances with feature names
features = ['List Price', 'NumPages', 'Thick']
importances = pd.DataFrame({'Feature': features, 'Importance': feature_importances})

# Sort by importance
importances = importances.sort_values(by='Importance', ascending=False)
print(importances)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Add a dummy column to use for hue (since barplot expects different hues)
importances['Feature_hue'] = importances['Feature']

# Plot the feature importances
plt.figure(figsize=(8, 6))
sns.barplot(
    data=importances, 
    x='Importance', 
    y='Feature', 
    hue='Feature_hue',  # Assign the 'Feature' column to hue
    palette='viridis', 
    dodge=False  # Ensure bars are not separated
)
plt.legend([], [], frameon=False)  # Disable the legend
plt.title('Feature Importances for Decision Tree Classifier')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
````

List price appears to be the most important feature for making predictions about Hard or Paperback books.
