The simple linear regression model predicts a dataset linearly. The independent variable is the predictor variable (x-axis) and the dependent variable is the outcome variable (y-axis). The linear regression line, with a slope and an intercept coefficient relevant to the model, is the line that best predicts the outcome variable based on the predictor variable, and minimizes error (difference between predicted Y value and actual Y value). Assuming that the errors around the regression line are normally distributed, a model can be created using this linear regression line and the normally distributed variance on either side of the line. Sampling from a normal distribution is relevant when sampling for predicted Y points that are based on the normally distributed error.

````python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm

# Step 1: Generate random X values (predictor variable)
np.random.seed(1)  # For reproducibility
n = 200  # Number of samples
X = np.linspace(0, 10, n)  # Predictor variable (evenly spaced values)

# Step 2: Define the slope and intercept
beta_0 = 5  # Intercept
beta_1 = 3  # Slope

# Step 3: Generate normally distributed errors
sigma = 5  # Standard deviation of the error
errors = norm.rvs(loc=0, scale=sigma, size=n)  # Normally distributed errors

# Step 4: Calculate the outcome variable Y
Y = beta_0 + beta_1 * X + errors  # Linear model with normal error term

# Step 5: Visualize the data
fig = go.Figure()

# Add scatter plot for generated data
fig.add_trace(go.Scatter(x=X, y=Y, mode='markers', name='Data'))

# Add theoretical regression line without any fitting
Y_true = beta_0 + beta_1 * X  # Regression line (without error term)
fig.add_trace(go.Scatter(x=X, y=Y_true, mode='lines', name='True Regression Line', line=dict(color='red')))

# Customize the layout
fig.update_layout(title='Theoretical Simple Linear Regression Model',
                  xaxis_title='Predictor Variable (X)',
                  yaxis_title='Outcome Variable (Y)',
                  showlegend=True)

# Show the plot
fig.show(renderer='png')
````

fitted_model.params give you information on the slope and intercept of the regression line fitted for a given dataset. In the case of questions 1-3, the given dataset is simulated using the true, predetermined, Y<sub>i</sub> equation. fitted_model.params values ($\hat{\beta}_0,\hat{\beta}_1$) are *estimated* (shown with the hat) based on the simulated dataset that is in turn based on the actual ${\beta}_0$ and ${\beta}_1$ values. 

This estimated slope and intercept are then applied onto each x<sub>i</sub> in the dataset, to calculate $\hat{y}_i$, which is what fittedvalues is. This process shows how fitted_model.fittedvalues are calculated, based on fitted_model.params, which is based off the fitted regression line for some dataset.

To use a visual explanation, please view the graph below. The blue dataset has been fitted to a linear regression model line. THat line is the orange dotted one. fitted_model.params describes the slope and intercept of that line. For each x value in the original dataset, fitted_model.fittedvalues basically plugs in the y = mx+b equation of the orange line, to get the green dots. These are the "fitted values".

````python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Sample dataset
np.random.seed(0)
X = np.random.rand(100, 3)
X = sm.add_constant(X)  # Add intercept
y = X @ np.array([1.5, -2.0, 3.0, 0.5]) + np.random.randn(100)

# Fit a regression model
fitted_model = sm.OLS(y, X).fit()

# Extract model parameters
params = fitted_model.params

# Manually calculate fitted values
fitted_values_manual = X @ params

# Fitted values from the model
fitted_values_model = fitted_model.fittedvalues

# Create the plot
plt.figure(figsize=(10, 6))

# Plot original y values
plt.scatter(fitted_values_model, y, color='dodgerblue', alpha=0.6, label="Original Values")

# Plot model fitted values
plt.scatter(fitted_values_model, fitted_values_manual, color='teal', alpha=0.6, label="Model Fitted Values (fitted_model.fittedvalues)")

# Add y = x line for reference
plt.plot([min(fitted_values_model), max(fitted_values_model)],
         [min(fitted_values_model), max(fitted_values_model)],
         color='darkorange', linestyle="--", linewidth=2, label="Fitted Model Line (fitted_model.params)")

# Draw lines between original values and fitted values from the model
for i in range(len(y)):
    plt.plot([fitted_values_model[i], fitted_values_model[i]], [fitted_values_model[i], y[i]], 
             color='gray', linestyle="--", linewidth=0.8, alpha=0.5)
    
# Labeling the plot
plt.xlabel("Fitted Values from Model")
plt.ylabel("Values")
plt.title("Comparison of Fitted Values with Original Data")
plt.legend()
plt.grid(True)

plt.show()
````

The ordinary least squares method calculates the residuals, which are the differences between $\hat{y}_i$ and $Y_i$. Basically, this is the difference between the observed y value and the predicted y value from the fitted model. This difference, squared, is the squared residual.
````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.utils import resample
import matplotlib.pyplot as plt
import seaborn as sns

# Applying the filter based on long_wait_limit and using the filtered data for analysis
long_wait_limit = 71
long_wait = old_faithful['waiting'] > long_wait_limit
filtered_data = old_faithful[long_wait]  # Filtered dataset

# Step 1: Bootstrap sampling for slope coefficients (using filtered data)
num_bootstrap_samples = 1000
bootstrap_slopes = []

for _ in range(num_bootstrap_samples):
    sample_data = resample(filtered_data)
    model = smf.ols('duration ~ waiting', data=sample_data).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Plotting bootstrapped slope coefficients
plt.figure(figsize=(10, 6))
sns.histplot(bootstrap_slopes, kde=True, color='purple', label='Bootstrap Slopes')
plt.axvline(np.mean(bootstrap_slopes), color='red', linestyle='--', label='Mean of Bootstrapped Slopes')
plt.title("Bootstrapped Sampling Distribution of Slope Coefficients")
plt.xlabel("Slope Coefficient")
plt.legend()
plt.show()

# Step 2: Simulate data under null hypothesis and collect slope coefficients (using filtered waiting values)
null_slopes = []
beta_0, beta_1, sigma = 1.65, 0, 0.37

for _ in range(num_bootstrap_samples):
    y_null = beta_0 + beta_1 * filtered_data['waiting'] + np.random.normal(0, sigma, len(filtered_data))
    null_data = pd.DataFrame({'waiting': filtered_data['waiting'], 'duration': y_null})
    model_null = smf.ols('duration ~ waiting', data=null_data).fit()
    null_slopes.append(model_null.params['waiting'])

# Plotting slope distribution under null hypothesis
plt.figure(figsize=(10, 6))
sns.histplot(null_slopes, kde=True, color='orange', label='Null Hypothesis Slopes')
plt.axvline(0, color='red', linestyle='--', label='True Null Slope (0)')
plt.title("Sampling Distribution of Slope Coefficients under Null Hypothesis")
plt.xlabel("Slope Coefficient")
plt.legend()
plt.show()

# Step 3: 95% Confidence Interval and p-value comparison
ci_lower, ci_upper = np.percentile(bootstrap_slopes, [2.5, 97.5])
contains_zero = (ci_lower <= 0 <= ci_upper)
simulated_p_value = np.mean(np.array(bootstrap_slopes) <= 0)
observed_model = smf.ols('duration ~ waiting', data=filtered_data).fit()
observed_p_value = observed_model.pvalues['waiting']

(ci_lower, ci_upper, contains_zero, simulated_p_value, observed_p_value)
````

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.

````python
# permutation test
import numpy as np
import pandas as pd

# Assuming `old_faithful` DataFrame has columns `duration` and `kind`
# Step 1: Calculate the observed difference in means between "short" and "long" groups
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Step 2: Set up permutation test
n_permutations = 10000  # Number of permutations
permutation_diffs = []

for _ in range(n_permutations):
    # Shuffle labels (reassign 'kind' randomly)
    shuffled = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1, replace=False).values)
    
    # Calculate the mean difference for this permutation
    mean_diff = shuffled.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permutation_diffs.append(mean_diff)

# Step 3: Calculate p-value
permutation_diffs = np.array(permutation_diffs)
p_value = np.mean(np.abs(permutation_diffs) >= np.abs(observed_diff))

# Output results
print(f"Observed difference in means: {observed_diff}")
print(f"Permutation-based p-value: {p_value}")
````

````python
# 95% bootstrapped CI
import numpy as np
import pandas as pd

# Assuming `old_faithful` DataFrame has columns `duration` and `kind`
# Step 1: Define a function to calculate the difference in means
def mean_difference(group1, group2):
    return np.mean(group1) - np.mean(group2)

# Step 2: Calculate the observed difference in means between "short" and "long" groups
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Step 3: Set up bootstrap resampling
n_bootstrap = 10000  # Number of bootstrap samples
bootstrap_diffs = []

# Perform bootstrap resampling
for _ in range(n_bootstrap):
    # Resample with replacement from each group
    short_sample = old_faithful[old_faithful['kind'] == 'short'].sample(n=len(old_faithful[old_faithful['kind'] == 'short']), replace=True)
    long_sample = old_faithful[old_faithful['kind'] == 'long'].sample(n=len(old_faithful[old_faithful['kind'] == 'long']), replace=True)
    
    # Calculate the mean difference for this bootstrap sample
    bootstrap_diff = mean_difference(short_sample['duration'], long_sample['duration'])
    bootstrap_diffs.append(bootstrap_diff)

# Step 4: Calculate the 95% bootstrap confidence interval
bootstrap_diffs = np.array(bootstrap_diffs)
lower_bound = np.percentile(bootstrap_diffs, 2.5)
upper_bound = np.percentile(bootstrap_diffs, 97.5)

# Output results
print(f"Observed difference in means: {observed_diff}")
print(f"95% Bootstrap Confidence Interval: ({lower_bound}, {upper_bound})")
````

a. The sampling approach for the first simulation is similar to sampling under the null (short or long does not affect duration), where you're trying to see if the difference between duration in the short and long groups is happening purely by chance. In the second approach, you're applying bootstrapping (sampling the sample) each group, creating a confidence interval, and looking at whether it makes sense to reject the null based on where it falls relative to the interval.

b. Permutation tests and boostrapping do not already assume a model structure, whereas indicator based does (the "definition" for short and long were defined prior to sampling). All models are trying to determine whether the null hypothesis can be rejected, just in different ways.
