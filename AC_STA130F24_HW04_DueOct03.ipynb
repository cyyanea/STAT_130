````python
import pandas as pd
import numpy as np

# URL of the Sleep Health and Lifestyle Dataset
df = pd.read_csv("https://raw.githubusercontent.com/OludolapoAnalyst/Sleep-and-Lifestyle-Dataset/main/Sleep_health_and_lifestyle_dataset.csv")
df
````

````python
# The column of interest will be quality of sleep - placed on a numerical scale for this dataset
column_of_interest = 'Quality of Sleep'  # This can be adjusted based on the interest of the analyzer

# Extract the sample data from the main dataset
sample_data = df[column_of_interest].dropna()  # Remove any NaN values for accurate sampling

# Number of bootstrap samples - this value can be adjusted as necessary
num_bootstrap_samples = 1000

# Function to calculate a 95% bootstrap confidence interval for a population parameter
# Input dataset, number of bootstrap samples, desired statistic for bootstrapping (default is mean) and desired confidence interval
def bootstrap_confidence_interval(data, num_bootstrap_samples=1000, statistic=np.mean, ci=95):
    bootstrap_stats = np.array([
        statistic(np.random.choice(data, size=len(data), replace=True))  # Generate bootstrap samples
        for _ in range(num_bootstrap_samples)
    ])
    
    # Calculate percentiles for the confidence interval
    lower_percentile = (100 - ci) / 2  # Lower bound percentile based on desired CI 
    upper_percentile = 100 - lower_percentile  # Upper bound percentile based on desired CI
    lower_bound = np.percentile(bootstrap_stats, lower_percentile)  # Lower CI bound
    upper_bound = np.percentile(bootstrap_stats, upper_percentile)  # Upper CI bound
    
    return lower_bound, upper_bound  # Return the confidence interval

# Calculate the 95% CI for the mean of the specified column
mean_ci = bootstrap_confidence_interval(sample_data, num_bootstrap_samples, statistic=np.mean)
print("95% CI for the mean of:", mean_ci)

# Calculate the 95% CI for the median of the specified column
median_ci = bootstrap_confidence_interval(sample_data, num_bootstrap_samples, statistic=np.median)
print("95% CI for the median:", median_ci)

# Calculate the 95% bootstrap confidence interval for the standard deviation
std_ci = bootstrap_confidence_interval(sample_data, num_bootstrap_samples, statistic=np.std)
print("95% CI for the standard deviation of:", std_ci)
````

````python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("patient_health_scores.csv")

# Define the number of bootstrap samples
n_bootstrap = 1000

# Create a function for bootstrapping
def bootstrap_mean_difference(data, n_bootstrap=1000):
    bootstrapped_means = []
    for _ in range(n_bootstrap):
        # Resample with replacement
        sample = data.sample(frac=1, replace=True)
        # Calculate the mean difference between Final and Initial Health Scores
        mean_diff = sample['FinalHealthScore'].mean() - sample['InitialHealthScore'].mean()
        bootstrapped_means.append(mean_diff)
    return np.array(bootstrapped_means)

# Apply bootstrapping to estimate the mean difference
bootstrapped_differences = bootstrap_mean_difference(df, n_bootstrap=n_bootstrap)

# Calculate the confidence intervals (e.g., 95% confidence interval)
ci_lower = np.percentile(bootstrapped_differences, 2.5)
ci_upper = np.percentile(bootstrapped_differences, 97.5)

# Output the results
mean_diff = np.mean(bootstrapped_differences)
print(f"Bootstrapped Mean Difference Between Final and Initial Health Score: {mean_diff}")
print(f"95% Confidence Interval: ({ci_lower}, {ci_upper})")
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate the bootstrapped distribution if not already available
n_bootstrap = 1000
bootstrapped_differences = bootstrap_mean_difference(df, n_bootstrap=n_bootstrap)

# Calculate the confidence intervals (95% CI)
ci_lower = np.percentile(bootstrapped_differences, 2.5)
ci_upper = np.percentile(bootstrapped_differences, 97.5)

# Plot 1: Histogram of bootstrapped mean differences
plt.figure(figsize=(10, 6))
sns.histplot(bootstrapped_differences, kde=True, color='blue', bins=30)
plt.axvline(ci_lower, color='red', linestyle='dashed', linewidth=2, label=f'2.5% CI: {ci_lower:.2f}')
plt.axvline(ci_upper, color='red', linestyle='dashed', linewidth=2, label=f'97.5% CI: {ci_upper:.2f}')
plt.axvline(np.mean(bootstrapped_differences), color='green', linestyle='solid', linewidth=2, label=f'Mean: {np.mean(bootstrapped_differences):.2f}')
plt.title('Histogram of Bootstrapped Mean Differences Between Final and Initial Health Score', fontsize=16)
plt.xlabel('Mean Difference (Final - Initial)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.show()

# Plot 2: Density plot with confidence intervals
plt.figure(figsize=(10, 6))
sns.kdeplot(bootstrapped_differences, fill=True, color='blue')
plt.axvline(ci_lower, color='red', linestyle='dashed', linewidth=2, label=f'2.5% CI: {ci_lower:.2f}')
plt.axvline(ci_upper, color='red', linestyle='dashed', linewidth=2, label=f'97.5% CI: {ci_upper:.2f}')
plt.axvline(np.mean(bootstrapped_differences), color='green', linestyle='solid', linewidth=2, label=f'Mean: {np.mean(bootstrapped_differences):.2f}')
plt.title('Density Plot of Bootstrapped Mean Differences Between Final and Initial Health Score', fontsize=16)
plt.xlabel('Mean Difference (Final - Initial)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
````
